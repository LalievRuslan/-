```python
# Импортируем библиотеки
from sympy import *
from sympy.plotting import plot
from sympy.solvers.inequalities import solve_univariate_inequality
```

```python
# присваиваем переменным методы из sympy
x = Symbol('x')
y = Symbol('y')
```

```python
# присваеваем переменной 'f' нашу функцию
f = 5*x**2 + 10*x - 150
f
```

```python
#1 Определяем корни
x1, x2 = solve(f)
```

```python
print('Первый корень  = ', x1.evalf())
print('Второй корень  = ', x2.evalf())
```

```python
#2,3,5 Находим интервалы на которых функция Возрастает/Убывает и вершину
f2 = 10 * x + 10
fun1 = plot(f2,(x,-2,1), show=False)
fun1.show()
```

```python
#Вершина
extremum = solve(f2)
extremum 
```

```python
#Интервалы
print(f'Функция убывает от -∞  до {extremum } и Функция возрастает с {extremum } до +∞')
```
